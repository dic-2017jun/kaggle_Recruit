```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from mpl_toolkits.basemap import Basemap
from sklearn.cluster import KMeans
```

```python
st=pd.read_csv('../input/air_store_info.csv')
```

```python
st.head()
```

## ジャンル名をラベル化

```python
genre_list=st.groupby('air_genre_name').count().index
```

```python
def list_to_dict(_list):
    _dict={}
    for i,p in enumerate(_list):
        _dict.update({p:i})
    return _dict
```

```python
genre_dict=list_to_dict(genre_list)
```

## 経度・緯度をk-meansでカテゴライズ

```python
f,ax = plt.subplots(1,1,figsize=(10,10))

# 緯度経度で範囲を指定する
north = 47
south = 30
east = 150
west = 125

# 地図の表示
m = Basemap(projection='merc',llcrnrlat=south,urcrnrlat=north, llcrnrlon=west,urcrnrlon=east)
# 海岸線を引く
m.drawcoastlines() 
m.fillcontinents(color='lightgray',zorder=1)
m.scatter(st['longitude'].values,st['latitude'].values,marker='o',color='blue',zorder=999,latlon=True)
plt.show()
```

```python
ll=st.drop(['air_store_id','air_genre_name','air_area_name'],axis=1)
```

```python
n_cl=12
km = KMeans(n_clusters=n_cl, init='random', n_init=10, max_iter=300, tol=1e-04, random_state=0)
ll_km = km.fit_predict(ll)
cla = [c[0] for c in km.cluster_centers_]
clo = [c[1] for c in km.cluster_centers_]
```

```python
lll=ll.copy()
lll['label']=ll_km
```

```python
lll.head()
```

```python
f,ax = plt.subplots(1,1,figsize=(10,10))

# 緯度経度で範囲を指定する
north = 47
south = 30
east = 150
west = 125

# 地図の表示
m = Basemap(projection='merc',llcrnrlat=south,urcrnrlat=north, llcrnrlon=west,urcrnrlon=east)
# 海岸線を引く
m.drawcoastlines() 
m.fillcontinents(color='lightgray',zorder=1)

cm = plt.get_cmap('gist_rainbow')
colors = [cm(2.*i/n_cl) for i in range(n_cl)]
colored=[colors[k] for k in lll['label']]
m.scatter(lll['longitude'].values,lll['latitude'].values,marker='o',color=colored,zorder=999,latlon=True)

m.scatter(clo,cla,color='Black',s=50,alpha=1,latlon=True,zorder=9999)
plt.setp(ax.get_yticklabels(), visible=True)
for i in range(len(cla)):
    xpt,ypt = m(clo[i],cla[i])
    plt.annotate(i, (xpt+500,ypt+500),zorder=99999,fontsize=16)
plt.show()

plt.show()
```

## ジャンルと緯度・経度情報を合体

```python
st1=st.copy()
st1['loc']=lll['label']
```

```python
st2=st1.replace({st1.columns[1]:genre_dict})
st2.drop(['air_area_name','latitude','longitude'],axis=1,inplace=True)
st2.head()
```

```python
st2.set_index('air_store_id',drop=True,inplace=True)
st2.to_csv('../dataset/genre_loc.csv')
```
