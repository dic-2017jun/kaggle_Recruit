```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from datetime import datetime as dt
```

```python
sample=pd.read_csv('../input/sample_submission.csv')
res=pd.read_csv('../dataset/reserve.csv')
gl=pd.read_csv('../dataset/genre_loc.csv')
date=pd.read_csv('../dataset/date.csv')
vis=pd.read_csv('../dataset/actual_visitors.csv')
tr=pd.read_csv('../dataset/train_data1.csv')
```

```python
print(sample.shape)
sample.head()
```

```python
df=sample.copy()
df=df.rename(columns={'id':'store_id_date'})
df['air_store_id']=[i[:i.find('_201')] for i in df['store_id_date']]
df['visit_date']=[i[-10:] for i in df['store_id_date']]
```

```python
df.head()
```

```python
date.head()
```

```python
#日付データのマージ
df1=pd.merge(df, date, on='visit_date', how='left')
```

```python
print(df1.shape)
df1.head()
```

```python
#genre,locationデータのマージ
df2=pd.merge(df1,gl,on='air_store_id',how='left')
```

```python
print(df2.shape)
df2.head()
```

```python
res.head()
```

```python
res1=res.drop(['air_store_id','visit_date'],axis=1)
df3=pd.merge(df2, res1, on='store_id_date', how='left')
```

```python
print(df3.shape)
df3.head()
```

```python
df4=df3.drop(['visit_date'],axis=1)
```

```python
print(df4.shape)
df4.head()
```

```python
gr=['year','month','day','day_of_week','working_day','holiday_flg','season']
tr_list=[]
for g in gr:
    col=['air_store_id',g,'mean_'+g,'max_'+g,'median_'+g]
    buf=pd.DataFrame(columns=col)
    for i in col:
        buf[i]=tr[i]
    buf=buf.drop_duplicates()
    tr_list.append(buf.reset_index(drop='True'))
```

```python
df5=df4.copy()
for i in range(len(gr)):
    df5=pd.merge(df5,tr_list[i],on=['air_store_id',gr[i]],how='left')
df5=df5.fillna(0)
```

```python
print(df5.shape)
df5.head()
```

```python
#列の順序入れ替え
col=['store_id_date',
    'year',
    'month',
    'day',
    'season',
    'day_of_week',
    'working_day',
    'holiday_flg',
    'air_genre_name',
    'air_area_name',
    'latitude',
    'longitude',
    'mean_year',
    'mean_month',
    'mean_day',
    'mean_day_of_week',
    'mean_working_day',
    'mean_holiday_flg',
    'mean_season',
    'max_year',
    'max_month',
    'max_day',
    'max_day_of_week',
    'max_working_day',
    'max_holiday_flg',
    'max_season',
    'median_year',
    'median_month',
    'median_day',
    'median_day_of_week',
    'median_working_day',
    'median_holiday_flg',
    'median_season',
    'visitors']

df6=pd.DataFrame(columns=col)
for c in col:
    df6[c]=df5[c]
```

```python
df6.set_index('store_id_date',drop=True,inplace=True)
df6.to_csv('../dataset/pre_submission.csv')
```
