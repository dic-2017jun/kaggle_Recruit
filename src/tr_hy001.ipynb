```python
#インポート
import numpy as np
import pandas as pd
import glob
import os
import datetime as dt
#ファイル読み込み
data = {
    #作成済みtestならびにtrainデータの読み込み
    'test':
    pd.read_csv('./traintest/test.csv'),
    'train':
    pd.read_csv('./traintest/train.csv'),
    #air_store_info_with_nearest_active_station.csvの読み込み
    'air_station':
    pd.read_csv('./weather/air_store_info_with_nearest_active_station.csv')
}
#ファイル結合
data['updated_train'] = pd.merge(data['train'], data['air_station'], how='inner', on=['air_store_id'])
data['updated_test'] = pd.merge(data['test'], data['air_station'], how='inner', on=['air_store_id'])

#気象データ読み込み(path変数は、個人毎に変更する必要有り※
path = 'C:/Users/hayyoshi/Desktop/Python/recruit/weather/1-1-16_5-31-17_Weather/*'
files = []
files = glob.glob(path)
path = 'C:/Users/hayyoshi/Desktop/Python/recruit/weather/1-1-16_5-31-17_Weather/'
files = []
for x in os.listdir(path):
    if os.path.isfile(path + x):  #isdirの代わりにisfileを使う
        files.append(x) 
for i in files:
    data.update({i.rstrip(".csv") :
    pd.read_csv('C:/Users/hayyoshi/Desktop/Python/recruit/weather/1-1-16_5-31-17_Weather/'+i, parse_dates = [2])}   
)
#気象データのカラムをもった、ゼロのデータフレーム作成
additional_columns=["avg_temperature","high_temperature","low_temperature",
                                                     "precipitation", "hours_sunlight","solar_radiation",
                                                     "deepest_snowfall","total_snowfall", "avg_wind_speed",
                                                     "avg_vapor_pressure","avg_local_pressure","avg_humidity",
                                                     "avg_sea_pressure","cloud_cover"]

df2 = pd.DataFrame(np.zeros([1, 14]), columns=additional_columns)
df2[:] = np.nan

#気象データのカラムを追加
data['updated_train'] = data['updated_train'].join(df2)
data['updated_test'] = data['updated_test'].join(df2)
data['updated_train'] = data['updated_train'].rename(columns = {'visit_date':'calendar_date'})
data['updated_test'] = data['updated_test'].rename(columns = {'visit_date':'calendar_date'})

train_1 = data["updated_train"]
test_1 = data["updated_test"]
temp_data_frame = pd.DataFrame(index=[], columns=["calendar_date","avg_temperature","high_temperature","low_temperature","precipitation","hours_sunlight","solar_radiation","deepest_snowfall","total_snowfall","avg_wind_speed","avg_vapor_pressure","avg_local_pressure","avg_humidity","avg_sea_pressure","cloud_cover"])

for a,b in zip(train_1["station_id"],train_1["calendar_date"]):
    temp = data[a]
    temp_data_frame = temp_data_frame.append(temp.ix[temp["calendar_date"] == b], ignore_index = True)
    train_1["avg_temperature"]=temp_data_frame["avg_temperature"]
    train_1["high_temperature"]=temp_data_frame["high_temperature"]
    train_1["precipitation"]=temp_data_frame["precipitation"]
    train_1["hours_sunlight"]=temp_data_frame["hours_sunlight"]
    train_1["solar_radiation"]=temp_data_frame["solar_radiation"]
    train_1["deepest_snowfall"]=temp_data_frame["deepest_snowfall"]
    train_1["total_snowfall"]=temp_data_frame["total_snowfall"]
    train_1["avg_wind_speed"]=temp_data_frame["avg_wind_speed"]
    train_1["avg_vapor_pressure"]=temp_data_frame["avg_vapor_pressure"]
    train_1["avg_local_pressure"]=temp_data_frame["avg_local_pressure"]
    train_1["avg_humidity"]=temp_data_frame["avg_humidity"]
    train_1["avg_sea_pressure"]=temp_data_frame["avg_sea_pressure"]
    train_1["cloud_cover"]=temp_data_frame["cloud_cover"]
    
for a,b in zip(test_1["station_id"],test_1["calendar_date"]):
    temp = data[a]
    temp_data_frame = temp_data_frame.append(temp.ix[temp["calendar_date"] == b], ignore_index = True)
    test_1["avg_temperature"]=temp_data_frame["avg_temperature"]
    test_1["high_temperature"]=temp_data_frame["high_temperature"]
    test_1["precipitation"]=temp_data_frame["precipitation"]
    test_1["hours_sunlight"]=temp_data_frame["hours_sunlight"]
    test_1["solar_radiation"]=temp_data_frame["solar_radiation"]
    test_1["deepest_snowfall"]=temp_data_frame["deepest_snowfall"]
    test_1["total_snowfall"]=temp_data_frame["total_snowfall"]
    test_1["avg_wind_speed"]=temp_data_frame["avg_wind_speed"]
    test_1["avg_vapor_pressure"]=temp_data_frame["avg_vapor_pressure"]
    test_1["avg_local_pressure"]=temp_data_frame["avg_local_pressure"]
    test_1["avg_humidity"]=temp_data_frame["avg_humidity"]
    test_1["avg_sea_pressure"]=temp_data_frame["avg_sea_pressure"]
    test_1["cloud_cover"]=temp_data_frame["cloud_cover"]
    
#ローカル環境に結果を出力
data['updated_train'].to_csv("updated_train20180113.csv")
data['updated_test'].to_csv("updated_test20180113.csv")
```
