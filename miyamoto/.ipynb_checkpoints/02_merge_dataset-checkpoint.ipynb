```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import preprocessing
from datetime import datetime as dt
```

```python
res=pd.read_csv('../dataset/reserve.csv')
gl=pd.read_csv('../dataset/genre_loc.csv')
date=pd.read_csv('../dataset/date.csv')
vis=pd.read_csv('../dataset/actual_visitors.csv')
```

```python
res.head()
```

```python
#idをid_dateの形に変形
res1=res.copy()
res1['store_id']=[i[:i.find('_201')] for i in res['store_id_date']]
res1['visit_date']=[i[-10:] for i in res['store_id_date']]
```

```python
res1.head()
```

```python
gl.head()
```

```python
#genre_locationとreserveをstore_idでマージ
df=pd.merge(res1, gl, on='store_id', how='left')
```

```python
df.head()
```

```python
date.head()
```

```python
#dateとdfをcalendar_dateでマージ
df1=pd.merge(df, date, on='visit_date', how='left')
```

```python
df1.head()
```

```python
vis.head()
```

```python
#visとdf1をstore_id_dateでマージ
df2=pd.merge(df1, vis, on='store_id_date', how='left')
```

```python
df2.head()
```

```python
#列の順序入れ替え
col=['store_id_date','store_id','year','month','day','season','day_of_week','working_day','holiday_flg','genre_name','loc','reserve_visitors','visitors']
df3=pd.DataFrame(columns=col)
for c in col:
    df3[c]=df2[c]
```

```python
df3.set_index('store_id_date',drop=True,inplace=True)
df3.to_csv('../dataset/train_data.csv')
```
