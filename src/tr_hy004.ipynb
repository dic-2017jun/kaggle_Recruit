```python
#インポート
import numpy as np
import pandas as pd
import glob
import os
import datetime as dt

#ファイル読み込み
tr = pd.read_csv('./traintest/tr_base.csv')
te = pd.read_csv('./traintest/te_base.csv')

#ストアIDカラム作成
tr_air_store_id=[]
for i in tr['id']:
    a,b,c=i.split('_')
    tr_air_store_id.append(a+'_'+b)
tr['air_store_id']=tr_air_store_id

te_air_store_id=[]
for j in te['id']:
    a,b,c=j.split('_')
    te_air_store_id.append(a+'_'+b)
te['air_store_id']=te_air_store_id
```

```python
#air_store_id毎のair_genreのvisitorsのmax,min,med,meanを算出
tr_air_genre_mean_src=tr.groupby(['air_store_id', 'air_genre_name'], as_index=False)['visitors'].mean()
tr_air_genre_median_src=tr.groupby(['air_store_id', 'air_genre_name'], as_index=False)['visitors'].median()
tr_air_genre_max_src=tr.groupby(['air_store_id', 'air_genre_name'], as_index=False)['visitors'].max()
tr_air_genre_min_src=tr.groupby(['air_store_id', 'air_genre_name'], as_index=False)['visitors'].min()
```

```python
#air_store_id毎のair_area_nameのvisitorsのmax,min,med,meanを算出
tr_air_area_mean_src=tr.groupby(['air_store_id', 'air_area_name'], as_index=False)['visitors'].mean()
tr_air_area_median_src=tr.groupby(['air_store_id', 'air_area_name'], as_index=False)['visitors'].median()
tr_air_area_max_src=tr.groupby(['air_store_id', 'air_area_name'], as_index=False)['visitors'].max()
tr_air_area_min_src=tr.groupby(['air_store_id', 'air_area_name'], as_index=False)['visitors'].min()

tr_air_area_mean_src
```

```python
#算出したair_store_id毎のair_genre_nameのvisitorsのmax,min,med,meanをマージ
for k in tr_air_genre_mean_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_genre_name'].values == k[1]), 'air_genre_name_visitor_mean'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_genre_name'].values == k[1]), 'air_genre_name_visitor_mean'] = k[2]
for k in tr_air_genre_median_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_genre_name'].values == k[1]), 'air_genre_name_visitor_median'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_genre_name'].values == k[1]), 'air_genre_name_visitor_median'] = k[2]
for k in tr_air_genre_max_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_genre_name'].values == k[1]), 'air_genre_name_visitor_max'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_genre_name'].values == k[1]), 'air_genre_name_visitor_max'] = k[2]
for k in tr_air_genre_min_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_genre_name'].values == k[1]), 'air_genre_name_visitor_min'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_genre_name'].values == k[1]), 'air_genre_name_visitor_min'] = k[2]

tr
```

```python
#算出したair_store_id毎のair_areaのvisitorsのmax,min,med,meanをマージ
for k in tr_air_area_mean_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_area_name'].values == k[1]), 'air_area_name_visitor_mean'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_area_name'].values == k[1]), 'air_area_name_visitor_mean'] = k[2]
for k in tr_air_area_median_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_area_name'].values == k[1]), 'air_area_name_visitor_median'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_area_name'].values == k[1]), 'air_area_name_visitor_median'] = k[2]
for k in tr_air_area_max_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_area_name'].values == k[1]), 'air_area_name_visitor_max'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_area_name'].values == k[1]), 'air_area_name_visitor_max'] = k[2]
for k in tr_air_area_min_src.values:
    tr.loc[(tr['air_store_id'].values == k[0])&(tr['air_area_name'].values == k[1]), 'air_area_name_visitor_min'] = k[2]
    te.loc[(te['air_store_id'].values == k[0])&(te['air_area_name'].values == k[1]), 'air_area_name_visitor_min'] = k[2]
```

```python
#算出したair_store_id毎のair_areaのvisitorsのmax,min,med,mean以外に追加した特徴量を削除
drop_col=["air_store_id"]
tr=tr.drop(drop_col,axis=1)
te=te.drop(drop_col,axis=1)
```

```python
#ファイル出力
tr.to_csv("tr_hy004.csv")
te.to_csv("te_hy004.csv")
```
